```python
#importing libraries
import numpy as np # linear algebra
import pandas as pd # plying with dataframe
import matplotlib.pyplot as plt
import operator
import seaborn as sns
```

```python
data=pd.read_csv("iris.csv")
data.head()
```

```python
print(data.info())
print(data['Name'].unique())
print(data.describe())
```

```python
plt.hist(data["SepalLength"], color = 'r')
plt.title('Sepal Length')
plt.xlabel('Length(cm)') 
plt.ylabel('Count')
plt.show()

plt.hist(data["SepalWidth"], color = 'r')
plt.title('Sepal Width')
plt.xlabel('Width(cm)') 
plt.ylabel('Count')
plt.show()

plt.hist(data["PetalLength"])
plt.title('Petal Length')
plt.xlabel('Length(cm)') 
plt.ylabel('Count')
plt.show()

plt.hist(data["PetalWidth"])
plt.title('Petal Width')
plt.xlabel('Width(cm)') 
plt.ylabel('Count')
plt.show()
```

```python
plt.boxplot([data["SepalLength"], data["SepalWidth"], data["PetalLength"], data["PetalWidth"]], 
            labels = ['Sepal Length', "Sepal Width", "Petal Length", "Petal Width"])
plt.show()
```
